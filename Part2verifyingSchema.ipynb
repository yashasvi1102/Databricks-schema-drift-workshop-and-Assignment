```python
# Check schema
spark.table("demo_cust_silver_sd").printSchema()

# Show sample rows with CreditScore
spark.table("demo_cust_silver_sd") \
    .select("CustomerID", "CreditScore", "signupDate", "source_filename") \
    .show(50, truncate=False)

# Check rescued data in bronze (optional)
spark.table("demo_cust_bronze_sd") \
    .select("source_filename", "_rescued_data") \
    .show(50, truncate=False)

# Row counts
from pyspark.sql.functions import *
df = spark.table("demo_cust_silver_sd")
df.select(
    count("*").alias("rows"),
    count(col("CreditScore")).alias("credit_present")
).show()
```
